```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes (valores arbitrarios para visualización)
C = 1.0
l = 1.0  # Momento angular
m = 1.0  # Masa del ión

# Definir la función de energía potencial efectiva
def V_eff(r):
    return -C/r**4 + (l**2)/(2*m*r**2)

# Rango de valores de r (evitando r=0)
r = np.linspace(0.1, 2.5, 500)
V = V_eff(r)

# Encontrar el máximo V0
r0 = np.sqrt(2*C*m/(l**2))  # Derivado de dV_eff/dr = 0
V0 = V_eff(r0)

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(r, V, 'b-', linewidth=2, label=r'$V_{\text{eff}}(r) = -\frac{C}{r^4} + \frac{\ell^2}{2m r^2}$')
plt.axhline(y=0, color='k', linestyle='--', linewidth=0.5)  # Eje horizontal
plt.axvline(x=r0, color='r', linestyle=':', linewidth=1, label=r'$r_0$')
plt.scatter(r0, V0, color='red', s=100, label=r'$V_0$')
plt.title('Energía Potencial Efectiva', fontsize=14)
plt.xlabel('Distancia $r$', fontsize=12)
plt.ylabel('$V_{\text{eff}}(r)$', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(-5, 2)  # Ajustar según sea necesario
plt.xlim(0, 2.5)
plt.show()
```
