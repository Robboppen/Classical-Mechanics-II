```python
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo de trayectoria y(x)
x = np.linspace(0, 5, 100)
y = np.log(x + 1)  # Trayectoria arbitraria

# Punto P y recta tangente
x_p = 2
y_p = np.log(x_p + 1)
y_prime = 1 / (x_p + 1)  # Derivada dy/dx en P

# Recta tangente en P
tangent_line = y_prime * (x - x_p) + y_p

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Trayectoria $y(x)$', color='blue')
plt.scatter(x_p, y_p, color='red', label='Punto $P$')
plt.plot(x, tangent_line, '--', color='green', label='Recta tangente en $P$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True)
plt.title('Relación geométrica entre $y\'$ y $\\theta$')
plt.show()

theta = np.linspace(0, np.pi/2 - 0.01, 100)  # Evita división por cero
y_prime = np.tan(theta)
sin_theta = np.sin(theta)
relation = y_prime / np.sqrt(1 + y_prime**2)

plt.figure(figsize=(8, 6))
plt.plot(theta, sin_theta, label='$\\sin \\theta$', color='red')
plt.plot(theta, relation, '--', label='$y\' / \\sqrt{1 + y\'^2}$', color='blue')
plt.xlabel('Ángulo $\\theta$ (radianes)')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.title('Demostración de $\\frac{y\'}{\\sqrt{1 + y\'^2}} = \\sin \\theta$')
plt.show()
```
