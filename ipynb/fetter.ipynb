```python
# Corregido de nuevo: usar \frac{1}{2} explícito
import numpy as np
import matplotlib.pyplot as plt

m = 1.0
k = 1.0
r = np.linspace(0.01, 5, 2000)  # evitar r=0

ls = [0.5, 1.0, 2.0]  # distintos momentos angulares a mostrar

def V_eff(r, l):
    return 0.5 * k * r**2 + (l**2) / (2 * m * r**2)

plt.figure(figsize=(8,6))
for l in ls:
    plt.plot(r, V_eff(r, l), label=f"l = {l}")

# Marcar Emin para l = 1.0 y mostrar puntos de inversión para E = 1.5 Emin
l0 = 1.0
Emin = np.sqrt(k * l0**2 / m)
E = 1.5 * Emin
# resolver r^2 = (E ± sqrt(E^2 - Emin^2)) / k
sqrt_term = np.sqrt(E**2 - Emin**2)
r2_plus = (E + sqrt_term) / k
r2_minus = (E - sqrt_term) / k
r_min = np.sqrt(r2_minus)
r_max = np.sqrt(r2_plus)

plt.hlines(E, xmin=0, xmax=5, linestyles='dashed', label=f"E = 1.5 Emin (l={l0})")
plt.vlines([r_min, r_max], ymin=0, ymax=E, linestyles='dashdot', label=f"puntos inversión (r_min,r_max) = ({r_min:.3f},{r_max:.3f})")

# marcar Emin en el eje
r0 = np.sqrt(l0/np.sqrt(m*k))
plt.scatter([r0], [Emin], marker='o', s=40)
plt.text(r0, Emin, f"  Emin={Emin:.3f}", va='bottom')

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel('r')
plt.ylabel('V_eff(r)')
plt.title(r'$V_{\mathrm{eff}}(r)=\frac{1}{2}k r^2 + \frac{l^2}{2 m r^2}$' )
plt.legend(loc='upper right', fontsize='small')
plt.grid(True)

plt.show()
```
