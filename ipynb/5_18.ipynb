```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros constantes
a = 1.0  # Valor arbitrario para 'a'
xi = np.pi / 4  # Ejemplo: ξ = π/4 (45°)

# Valores constantes para q1 y q2 (ejemplo)
q1_const = 0.5
q2_const = -0.5

# Rango del ángulo ϕ
phi = np.linspace(0, 4 * np.pi, 1000)  # De 0 a 4π

# Curva para q1 = constante
r_q1 = a * np.exp(q1_const) * np.exp(phi * (1 / np.tan(xi)))  # cot(ξ) = 1/tan(ξ)

# Curva para q2 = constante
r_q2 = a * np.exp(q2_const) * np.exp(-phi * np.tan(xi))

# Convertir a coordenadas cartesianas para graficar
x_q1 = r_q1 * np.cos(phi)
y_q1 = r_q1 * np.sin(phi)

x_q2 = r_q2 * np.cos(phi)
y_q2 = r_q2 * np.sin(phi)

# Graficar
plt.figure(figsize=(8, 8))
plt.plot(x_q1, y_q1, label=f'$q_1 = {q1_const}$ (Espiral creciente)')
plt.plot(x_q2, y_q2, label=f'$q_2 = {q2_const}$ (Espiral decreciente)', linestyle='--')

# Añadir detalles
plt.title('Curvas de $q_1$ y $q_2$ constantes en coordenadas polares')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.axis('equal')  # Misma escala en ejes X e Y
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros ajustados para visualización clara
a = 1.0
xi = np.pi / 6  # ξ = 30° (ángulo más pequeño para espirales más suaves)
q1_const = 0.0  # q1 constante (puedes ajustarlo luego)
q2_const = 0.0  # q2 constante (puedes ajustarlo luego)

# Rango de ϕ controlado para evitar divergencias
phi = np.linspace(0, 10, 1000)  # Rango reducido para evitar crecimiento exponencial excesivo

# Ecuaciones de las espirales (evitando divergencia)
r_q1 = a * np.exp(q1_const + phi * (1 / np.tan(xi)))  # q1: espiral creciente
r_q2 = a * np.exp(q2_const - phi * np.tan(xi))        # q2: espiral decreciente

# Convertir a coordenadas cartesianas
x_q1, y_q1 = r_q1 * np.cos(phi), r_q1 * np.sin(phi)
x_q2, y_q2 = r_q2 * np.cos(phi), r_q2 * np.sin(phi)

# Graficar
plt.figure(figsize=(10, 10))
plt.plot(x_q1, y_q1, label=f'$q_1 = {q1_const}$ (Creciente)', color='blue', linewidth=2)
plt.plot(x_q2, y_q2, label=f'$q_2 = {q2_const}$ (Decreciente)', color='red', linestyle='--', linewidth=2)

# Ajustar límites para mostrar ambas espirales
max_r = max(np.max(r_q1), np.max(r_q2))
plt.xlim(-max_r, max_r)
plt.ylim(-max_r, max_r)

# Detalles estéticos
plt.title('Espirales $q_1$ y $q_2$ constantes\n' + r'$\xi = \pi/6$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.axis('equal')  # Misma escala en ejes X e Y
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros fijos
a = 1.0
xi = np.pi / 4  # ξ = 45° (ejemplo estándar)
q1_value = 0.5  # Valor constante para q1
q2_value = -0.5  # Valor constante para q2

# Rango del ángulo ϕ (ajustado para evitar escalas extremas)
phi = np.linspace(0, 4 * np.pi, 500)  # De 0 a 4π

# Ecuaciones directas de las curvas
r_q1 = a * np.exp(q1_value + phi * (1 / np.tan(xi)))  # q1 = constante (crece)
r_q2 = a * np.exp(q2_value - phi * np.tan(xi))          # q2 = constante (decrece)

# Conversión a coordenadas cartesianas
x_q1 = r_q1 * np.cos(phi)
y_q1 = r_q1 * np.sin(phi)
x_q2 = r_q2 * np.cos(phi)
y_q2 = r_q2 * np.sin(phi)

# Crear el gráfico
plt.figure(figsize=(8, 8))
plt.plot(x_q1, y_q1, label=f'$q_1 = {q1_value}$', color='blue')
plt.plot(x_q2, y_q2, label=f'$q_2 = {q2_value}$', color='red', linestyle='--')

# Ajustes estéticos
plt.title('Curvas de $q_1$ y $q_2$ constantes')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.axis('equal')  # Misma escala en ambos ejes
plt.tight_layout()
plt.show()
```
