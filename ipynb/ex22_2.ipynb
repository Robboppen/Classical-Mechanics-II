```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del problema
v0 = 10  # cm/s
tau = 5  # s

# Definimos la funcin de distancia recorrida s(t)
def s(t):
    t = np.array(t)
    s_values = np.zeros_like(t)
    for i, ti in enumerate(t):
        if ti <= tau:
            s_values[i] = v0 * (ti - ti**2 / (2 * tau))
        else:
            # Distancia hasta t=5s + distancia retrocedida
            s_forward = v0 * (tau - tau**2 / (2 * tau))  # = 25 cm
            s_back = np.abs(v0 * ((ti - ti**2 / (2 * tau)) - (tau - tau**2 / (2 * tau))))
            s_values[i] = s_forward + s_back
    return s_values

# Creamos un array de tiempos desde 0 hasta 20 segundos
t = np.linspace(0, 100, 500)


# Calculamos s(t)
s_t = s(t)

# Creamos la grfica
plt.figure(figsize=(10, 6))
plt.plot(t, s_t, 'b-', linewidth=2, label='Distancia recorrida s(t)')

# Marcamos los puntos importantes
plt.scatter([5], [25], color='red', zorder=5)
plt.text(5, 25, '  t=5s (25 cm)', verticalalignment='bottom')

plt.scatter([8], [34], color='green', zorder=5)
plt.text(8, 34, '  t=8s (34 cm)', verticalalignment='bottom')

# Configuramos el grfico
plt.title('Distancia recorrida s(t) en funcin del tiempo', fontsize=14)
plt.xlabel('Tiempo t (s)', fontsize=12)
plt.ylabel('Distancia s(t) (cm)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.xlim(0, 100)
plt.ylim(0, 100)

# Mostramos el grfico
plt.show()
```
