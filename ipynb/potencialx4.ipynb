```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
U0 = 1
x0 = 1
x = np.linspace(-2*x0, 2*x0, 500)
U = U0 * (-2*(x/x0)**2 + (x/x0)**4)

# Gráfica
plt.figure(figsize=(8,5))
plt.plot(x, U, label=r"$U(x)=U_0\left[-2\left(\frac{x}{x_0}\right)^2+\left(\frac{x}{x_0}\right)^4\right]$")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(-x0, color='red', linestyle='--', label=r"$x=-x_0$")
plt.axvline(0, color='green', linestyle='--', label=r"$x=0$")
plt.axvline(x0, color='red', linestyle='--', label=r"$x=+x_0$")
plt.xlabel(r"$x$")
plt.ylabel(r"$U(x)$")
plt.title("Potencial $U(x)$")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros (puedes cambiarlos)
U0 = 1.0
r0 = 1.0
m = 0.5  # masa efectiva (ejemplo). Si son dos átomos idénticos de masa M, usar m = M/2.

# Dominio evitando r=0
r = np.linspace(0.7*r0, 2.5*r0, 800)
U = U0 * ((r0/r)**12 - 2*(r0/r)**6)

# Cálculo de k y omega
k = 72 * U0 / r0**2
omega = np.sqrt(k / m)

print(f"k = {k:.4g}   omega = {omega:.4g}  (con m = {m})")

# Gráfico
plt.figure(figsize=(8,5))
plt.plot(r, U, label=r"$U(r)=U_0\left[(r_0/r)^{12}-2(r_0/r)^6\right]$")
plt.axvline(r0, color='red', linestyle='--', label=r"$r=r_0$ (mínimo)")
plt.scatter([r0],[U0*((r0/r0)**12-2*(r0/r0)**6)], color='red')
plt.annotate(r"$r_0$",
             xy=(r0, U0*((r0/r0)**12-2*(r0/r0)**6)),
             xytext=(1.05*r0, 0.2*U0),
             arrowprops=dict(arrowstyle="->", lw=0.8))
plt.xlabel(r"$r$")
plt.ylabel(r"$U(r)$")
plt.title("Potencial de Lennard–Jones (6–12)")
plt.grid(True)
plt.legend()
plt.ylim(min(U)*1.1, max(U)*0.9)
plt.show()
```
