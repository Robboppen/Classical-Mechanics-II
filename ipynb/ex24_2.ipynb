```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
g = 9.8
theta = np.arctan(1.5)  # 56.31°
v0 = 15
v0y = v0 * np.sin(theta)

# Tiempos
t_botella = np.linspace(0, 2.47, 100)  # Hasta que la botella toca el suelo
t_piedra = np.linspace(0, 3, 100)      # Rango extendido para visualización

# Ecuaciones
y_botella = 30 - 0.5 * g * t_botella**2
y_piedra = v0y * t_piedra - 0.5 * g * t_piedra**2

# Tiempo teórico de choque (no físico)
t_choque_teorico = 20 / (15 * np.cos(theta))

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(t_botella, y_botella, 'r-', label='Botella: $y_b(t) = 30 - 4.9 t^2$')
plt.plot(t_piedra, y_piedra, 'b-', label='Piedra: $y_p(t) = 12.49 t - 4.9 t^2$')
plt.scatter(t_choque_teorico, 30 - 4.9 * t_choque_teorico**2, color='blue', marker='*', s=200, label='Choque teórico (no físico)')
plt.axvline(x=2.47, color='gray', linestyle='--', label='Botella toca el suelo (t=2.47 s)')
plt.xlabel('Tiempo $t$ (s)')
plt.ylabel('Altura $y$ (m)')
plt.title('Trayectorias verticales de la botella y la piedra')
plt.legend()
plt.grid(True)
plt.ylim(0, 35)
plt.show()
```
